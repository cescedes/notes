```python
#max value
def max_value(nums):
    maximum = float('-inf')
    for num in nums:
        if num > maximum:
            maximum = num
    return maximum
```

```python
#prime number
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
#prime number opt
from math import sqrt, floor
def is_prime_opt(n):
    if n < 2:
        return False
    for i in range(2, floor(sqrt(n))+1):
        if n % i == 0:
            return False
    return True
```

time complexity for dictionary vs. list is different.
for dictionary key lookup you will get constant O(1), however lookup in a list will be linear O(n). dictionary is implemented as hash tables. in list you go through every member, which doesnt happen in dict except the worst case.

appending to list is constant. inserting into a list is linear. when you insert it goes to the top of the list and shifts the indexes of all elements.

```python
def anagrams(s1, s2):
    return char_count(s1) == char_count(s2)

def char_count(s):
    count = {}

    for char in s:
        if char not in count:
            count[char] = 0
        count[char]+=1
    return count

anagrams('cats', 'tacs')
```

```python
#another version with builtin packages
from collections import Counter
def anagrams(s1, s2):
    return Counter(s1) == Counter(s2)

anagrams('cats', 'tacs')
```

for anagrams the order is not important, only frequency.

```python
#most frequent char, if there are ties return the earliest
def most_freq_char(s):
    count = {}

    for char in s:
        if char not in count:
            count[char] = 0
        count[char]+=1
    
    most = None

    for char in s:
        if most is None or count[char] > count[most]:
            most = char

    return most

most_freq_char('catssss')
```

```python
#pair sum: brute force, O(n^2)
def pair_sum(nums, target):
    for i in range(0, len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] ==target:
                return(i,j)

#pair sum: hash map, O(n)
def pair_sum_hm(nums, target):
    prev_nums = {}
    for index, num in enumerate(nums): #enumarate for access to index
        complement = target - num
        if complement in prev_nums:
            return (prev_nums[complement], index)
    prev_nums[num] = index
```

```python
#pair product: hashmap, O(n)
def pair_product(nums, target):
    prev_nums = {}
    for index, num in enumerate(nums):
        complement = target / num
        if complement in prev_nums:
            return (index, prev_nums[complement])
    prev_nums[num] = index
```

```python
nums = [2, 3, 4, 5, 10, 15]
target = 15

pair_sum(nums, target)
```

```python
#intersection
def intersection(a, b):
    result = []
    for item in a:
        if item in b:
            result.append(item)
    return result
```

```python
def intersection_2(a,b):
    item_set = set(a)
    return [element for element in b if element in item_set]
```

```python
#sum numbers recursive (not a better solution, time and space O(n^2))
def sum_recursive(nums):
    if len(nums) == 0:
        return 0
    return nums[0] + sum_recursive(nums[1:])

sum_recursive(nums)
```

```python
#basic linked list traversal
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

a = Node('A')
b = Node('B')
c = Node('C')
d = Node('D')

a.next = b
b.next = c
c.next = d
```

```python
def print_list(head):
    current = head
    while current is not None:
        print(current.val)
        current = current.next

print_list(a)
```

```python
def print_list_recursion(head):
    if head is None:
        return
    print(head.val)
    print_list_recursion(head.next)

print_list_recursion(a)
```

```python
def list_values(head):
    current = head
    values = []
    while current is not None:
        values.append(current.val)
        current = current.next
    return(values)

list_values(a)
```

```python
def list_values_recursive(head):
    values = []
    fill_values(head, values)
    return values

def fill_values(head, values):
    if head is None:
        return
    values.append(head.val)
    fill_values(head.next, values)

list_values_recursive(a)
```

```python
def sum_linked_list(head):
    total_sum = 0
    current = head
    while current is not None:
        total_sum += current.val
        current = current.next

    return total_sum
```

```python
def sum_linked_list_recursive(head):
    if head is None:
        return 0
    return head.val + sum_linked_list_recursive(head.next)
```
